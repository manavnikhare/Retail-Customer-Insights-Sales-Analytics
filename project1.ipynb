```python
#import panda lib
import pandas as pd
```

```python
#import csv file
df = pd.read_csv('customer_shopping_behavior.csv')
```

```python
#statistics
df.describe(include='all')
```

```python
#filling the null values with median values
df['Review Rating'] = df.groupby('Category')['Review Rating'].transform(lambda x: x.fillna(x.median()))
```

```python
#checking for null values
pd.isnull(df).sum()
```

```python
#renaming names of column using underscore as it will make our analysis consistent in sql and python also renaming purchase amount
df.columns=df.columns.str.lower()
df.columns=df.columns.str.replace(' ','_')
df=df.rename(columns={'purchase_amount_(usd)':'purchase_amount'})
```

```python
df.columns
```

```python
#creating column age_groups based on the data distribution of age using qcut which will divide age distribution in 4 groups
labels=['Young-Adult','Adult','Middle-Aged','Senior']
df['age_group']=pd.qcut(df['age'],q=4, labels=labels)
```

```python
df[['age','age_group']].head(10)
```

```python
# create column purchase_frequency_days

frequency_mapping = {
    'Fortnightly': 14,
    'Weekly': 7,
    'Monthly': 30,
    'Quarterly': 90,
    'Bi-Weekly': 14,
    'Annually': 365,
    'Every 3 Months': 90
}

df['purchase_frequency_days'] = df['frequency_of_purchases'].map(frequency_mapping)
```

```python
#lets check our table
df[['frequency_of_purchases','purchase_frequency_days']].head(10)
```

```python
#check discount and promocode
df[['discount_applied','promo_code_used']].head(10)
```

```python
#check whether they are same throughout
(df['discount_applied']==df['promo_code_used']).all()
```

```python
#true aa rha hai which means same hai toh inki zarurat nhi hai hume 
df=df.drop('promo_code_used',axis=1)
```
